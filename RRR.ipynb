```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# Use a pre-trained model for embedding
embedder = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight, fast, accurate

# Compute document embeddings
doc_embeddings = embedder.encode(documents, convert_to_numpy=True)
```

```python
# Build the index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
def retrieve_relevant_chunks(query, top_k=2):
    query_embedding = embedder.encode([query], convert_to_numpy=True)
    scores, indices = index.search(query_embedding, top_k)
    return [documents[i] for i in indices[0]]
```

```python
from transformers import pipeline

# Load a small open-source generative model; for real use, pick any large language model ('distilbert', 'mistral', 'llama', or 'gpt2' for demo if resources limited)
qa_pipeline = pipeline("text-generation", model="gpt2")  # You can replace with a more capable LLM for better answers

def rag_answer(query):
    # 1. Retrieve
    context_chunks = retrieve_relevant_chunks(query)
    context = " ".join(context_chunks)

    # 2. Build the prompt for the LLM
    prompt = f"Context: {context}\nQuestion: {query}\nAnswer:"
    
    # 3. Generate an answer using the LLM
    result = qa_pipeline(prompt, max_length=100, num_return_sequences=1)
    answer = result[0]['generated_text'].split("Answer:")[-1].strip()
    return answer
```

```python
user_query = "What does RAG stand for?"
answer = rag_answer(user_query)
print("Question:", user_query)
print("Answer:", answer)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
from transformers import pipeline

# 1. Prepare Documents
documents = [
    "The capital of France is Paris.",
    "Python is a popular programming language.",
    "The Eiffel Tower is located in Paris.",
    "Machine learning is a subset of artificial intelligence.",
    "RAG stands for Retrieval-Augmented Generation."
]

# 2. Embed Documents
embedder = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = embedder.encode(documents, convert_to_numpy=True)

# 3. Build FAISS Index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)

# 4. Define Retriever
def retrieve(query, k=2):
    query_embedding = embedder.encode([query], convert_to_numpy=True)
    scores, indices = index.search(query_embedding, k)
    return [documents[i] for i in indices[0]]

# 5. Use QA Model for Generation
qa_pipeline = pipeline('question-answering', model='deepset/roberta-base-squad2')

def rag_answer(query):
    context_chunks = retrieve(query)
    # Combine all retrieved chunks for maximum context
    context = ' '.join(context_chunks)
    # Ask QA pipeline
    result = qa_pipeline({'question': query, 'context': context})
    return result['answer']

# 6. Test
user_query = "What does RAG stand for?"
answer = rag_answer(user_query)
print('Question:', user_query)
print('Answer:', answer)
```
